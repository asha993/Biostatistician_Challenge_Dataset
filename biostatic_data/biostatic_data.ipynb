```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("/content/Biostatistician_Challenge_Dataset.xlsx - biostatistician_assignment.csv")
df
```

```python
#checking data types of each columns
df.dtypes
```

```python
#checking missing values
df.isna().sum()
```

```python
#finding duplicate values
df.duplicated().sum()
```

```python
df['tobacco_consumption'] = pd.to_numeric(df['tobacco_consumption'], errors='coerce')
df['tobacco_consumption'] = df['tobacco_consumption'].fillna(0).astype(int)
df
```

```python
df.corr()
```

```python
#storing categorical column names to a new variable
categorical=[i for i in df.columns if df[i].dtype=='O']
categorical
```

```python
#for categorical values we can replace the null values with the Mode of it
for i in categorical:
    df[i].fillna(df[i].mode()[0],inplace=True)
```

```python
#storing int,float column names to a new variable
num=[i for i in df.columns if df[i].dtypes!='O']
num
```

```python
#for int,float values we can replace the null values with the mean of it
for i in num:
   df[i].fillna(df[i].mean(),inplace=True)
```

```python
#rechecking missing values
df.isna().sum()
```

```python
#applying label encoding for gender
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df.gender=le.fit_transform(df.gender)
df
```

```python
#here converting rest of the categorical datas using label encoding
#alcohol_consumption
#'is_literate',

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df.alcohol_consumption=le.fit_transform(df.alcohol_consumption)
df.is_literate=le.fit_transform(df.is_literate)
df
```

```python
#using label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df.primary_cooking_fuel=le.fit_transform(df.primary_cooking_fuel)
df
```

```python
#count plot for categorical values
plt.figure(figsize=(10,200))
plotnumber = 1

for col in categorical:
    if plotnumber <= df.shape[1] and col!='Pedestrian_movement':
        ax1 = plt.subplot(28,1,plotnumber)
        sns.countplot(data=df, y=col, palette='muted')
        plt.xticks(fontsize=12)
        plt.yticks(fontsize=12)
        plt.title(col.title(), fontsize=14)
        plt.xlabel('')
        plt.ylabel('')
    plotnumber +=1
```

```python
plt.figure(figsize=(22,17))
sns.set(font_scale=1)
sns.heatmap(df.corr(), annot=True)
```

```python
#distribution for numerical columns
plt.figure(figsize=(10,15))
plotnumber = 1
for i in num:
    if plotnumber <= df.shape[1]:
        ax1 = plt.subplot(4,3,plotnumber)
        plt.subplots_adjust(left=0.125, bottom=0.2  , right=0.9, top=0.9 , wspace=0.8, hspace=0.6)
        plt.hist(df[i],color='red')
        plt.xticks(fontsize=12)
        plt.yticks(fontsize=12)
        plt.title(' '+i, fontsize=10)
    plotnumber +=1
  
```

```python
#here converting education status string to numeric and removing special characters too
import re
df['education_status'] = [re.search('[0-9]', x)[0] for x in df['education_status']]
df
```

```python
#converting string to numeric
df['primary_transport'] = pd.to_numeric(df['primary_transport'],errors='coerce')
df = df.replace(np.nan, 0, regex=True)
df['primary_transport'] = df['primary_transport'].astype(int)
df
```

```python
#replace mixed values with NaN ,convert column to numeric
df['primary_house_material'] = pd.to_numeric(df['primary_house_material'],errors='coerce')
df = df.replace(np.nan, 0, regex=True)
df['primary_house_material'] = df['primary_house_material'].astype(int)
df = df.replace(np.nan, 0, regex=True)
df
```

```python
#getting data types
df.dtypes
```

```python
#converting floating to integer
df['sdh_occupational_risk']=df['sdh_occupational_risk'].astype(int)
df['age']=df['age'].astype(int)
df['occupation_status']=df['occupation_status'].astype(int)
df['primary_drinking_water']=df['primary_drinking_water'].astype(int)
df['primary_toilet']=df['primary_toilet'].astype(int)
df['toilet_usage_status']=df['toilet_usage_status'].astype(int)
df['primary_electricity']=df['primary_electricity'].astype(int)
df
```

```python
#converting object to integer
df['education_status'] = df['education_status'].astype(int)
df
```

```python

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
#creating model using Logistic regression 
from sklearn.linear_model import LogisticRegression
mod=LogisticRegression()
mod.fit(x_train,y_train)
y_pred=mod.predict(x_test)
y_pred
print(f"Model Accuracy: {mod.score(x_test, y_test)}")
```

```python
#creating model using linear regression 
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
#applying a value for prediction 
print(model.predict([[0,1,0,0,1,3,1,29,6,9,1,5,6,0,0,1,1,0]]))
```

```python
#ploting slop y=mx+c

print("slop is",model.coef_)
print("constant is",model.intercept_)
```
